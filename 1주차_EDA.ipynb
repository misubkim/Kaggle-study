## object features 확인

```python
features_object = [col for col in data if data[col].dtype == 'object']
features_object
```

dependency 변수값의 제곱이 SQBdependency 변수값과 같으므로  
SQBdependency 변수값의 sqrt값으로 dependency 변수값을 대체

```python
data['dependency'].unique()
```

```python
data['dependency'] = np.sqrt(data['SQBdependency'])
```

edjefe, edjefa 변수는 'education of male/female head of household'이므로  
yes값은 1로, no값은 0으로 대체

```python
data['edjefe'].unique()
```

```python
data['edjefa'].unique()
```

```python
def replace(x):
    if x == 'yes':
        return 1
    elif x == 'no':
        return 0
    else:
        return x
```

```python
data['edjefe'] = data['edjefe'].apply(replace).astype(float)
data['edjefa'] = data['edjefa'].apply(replace).astype(float)
```

새 변수로 'max education of either head of household'인 edjef 변수 추가

```python
data['edjef'] = np.max(data[['edjefa','edjefe']], axis=1)
```

## 변수 제거

binary 변수 정의

```python
binary_features = [col for col in data.columns if data[col].value_counts().shape[0] == 2]
```

continuous 변수 정의

```python
continuous_features = [col for col in data.columns if col not in binary_features]
continuous_features = [col for col in continuous_features if col not in features_object]
continuous_features = [col for col in continuous_features if col not in ['Target']]
```

```python
len(continuous_features)
```

hhsize 변수는 'household size'를, tamhog는 'size of the household'로 같은 값을 나타내므로  
tamhog 변수 제거

```python
sum(data['hhsize']!=data['tamhog'])
```

```python
data.drop('tamhog', axis=1, inplace=True)
```

Squared 변수 제거

```python
data = data[[x for x in data if not x.startswith('SQB')]]
data = data.drop(columns = ['agesq'])
data.shape
```

highly correlated 변수 제거

```python
corr_matrix = data.corr()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(abs(upper[column])>0.975)]
print(f'There are {len(to_drop)} correlated columns to remove.')
print(to_drop)
```

```python
data = data.drop(columns=to_drop)
data.shape
```

다시 binary, continuous features 정의

```python
binary_features = [col for col in data.columns if data[col].value_counts().shape[0] == 2]
```

```python
continuous_features = [col for col in data.columns if col not in binary_features]
continuous_features = [col for col in continuous_features if col not in features_object]
continuous_features = [col for col in continuous_features if col not in ['Target']]
```

```python
all_equal = df_train.groupby('idhogar')['Target'].apply(lambda x:x.nunique()==1)
not_equal = all_equal[all_equal != True]
print('There are {} households where the family members do not all have the same target'.format(len(not_equal)))
```

```python
train[train['idhogar'] == not_equal.index[0]][['idhogar','parentesco1','Target']]
```

같은 household 내 사람들은 같은 head of household를 가지도록 함
household의 head이면 parentesco1이 1이다. 

```python
households_leader = df_train.groupby('idhogar')['parentesco1'].sum()
households_no_head = df_train.loc[df_train['idhogar'].isin(households_leader[households_leader==0].index),:]
print('There are {} households without a head.'.format(households_no_head['idhogar'].nunique()))
```

```python
households_no_head_equal = households_no_head.groupby('idhogar')['Target'].apply(lambda x: x.nunique()==1)
print('{} households with no head have different labels.'.format(sum(households_no_head_equal==False)))
```

```python
for household in not_equal.index:
    true_target = int(df_train[(df_train['idhogar']==household) &
                              (train['parentesco1']==1.0)]['Target'])
    df_train.loc[df_train['idhogar']==household, 'Target']=true_target

all_equal = df_train.groupby('idhogar')['Target'].apply(lambda x:x.nunique()==1)
not_equal=all_equal[all_equal != True]
print('There are {} households where the family members do not all have the same target.'.format(len(not_equal)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Target과의 correlation

```python
from scipy.stats import spearmanr
import warnings
warnings.filterwarnings("ignore")

labels = []
values = []
for col in df_train.columns:
    if col not in ["Id", "Target"]:
        labels.append(col)
        values.append(spearmanr(df_train[col].values, df_train["Target"].values)[0])
corr_df = pd.DataFrame({'col_labels':labels, 'corr_values':values})
corr_df = corr_df.sort_values(by='corr_values')
 
corr_df = corr_df[(corr_df['corr_values']>0.1) | (corr_df['corr_values']<-0.1)]
ind = np.arange(corr_df.shape[0])
width = 0.9
fig, ax = plt.subplots(figsize=(12,30))
rects = ax.barh(ind, np.array(corr_df.corr_values.values), color='red')
ax.set_yticks(ind)
ax.set_yticklabels(corr_df.col_labels.values, rotation='horizontal')
ax.set_xlabel("Correlation coefficient")
ax.set_title("Correlation coefficient of the variables")
plt.show()
```
